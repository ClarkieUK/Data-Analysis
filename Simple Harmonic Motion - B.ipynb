```python
ypos = data['y']
time = data['t']
xpos = data['x']
```

```python
x = time
y = xpos


#find line of best fit
a,b,c,d,e = np.polyfit(x,y,4)
plt.plot(x, a*x**4 + b*x**3 + c*x**2 + d*x + e,color='orange',zorder=0)
plt
#add points to plot
plt.scatter(x,y,color='steelblue')

plt.title("The positional tracking of the oscillation of a pendulum")
plt.xlabel("Time (s)")
plt.ylabel("X Displacement from origin (m)")


plt.savefig("pendulumtrack.PNG")
plt.show()
```
